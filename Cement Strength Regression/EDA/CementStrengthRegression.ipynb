```python
plt.figure(figsize=(15,15),facecolor='white')
plotnumber = 1

for column in df.columns:
    ax = plt.subplot(4,3,plotnumber)
    sns.distplot(df[column])
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
X = df.drop(['Concrete compressive strength(MPa, megapascals) '], axis=1)
y = df['Concrete compressive strength(MPa, megapascals) ']
```

```python
for col in X.columns:
    X[col]+=1
    X[col] = np.log(X[col])
```

```python
plt.figure(figsize=(15,15),facecolor='white')
plotnumber = 1

for column in X.columns:
    ax = plt.subplot(4,3,plotnumber)
    sns.distplot(X[column])
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
X.head()
```

```python
plt.figure(figsize=(15,15),facecolor='white')
plotnumber = 1

for column in X.columns:
    ax = plt.subplot(4,3,plotnumber)
    sns.boxplot(X[column])
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
# let's check how our features are related to the target column
plt.figure(figsize=(20,20),facecolor='white')
plotnumber = 1

for column in X.columns:
    ax = plt.subplot(4,3,plotnumber)
    sns.scatterplot(X[column],y)
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(X.corr().abs(), annot=True, vmin=-0.5, vmax=1)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 1 / 3, random_state=36)
```

```python
x_train
```

```python
x_train = pd.DataFrame(scalar.fit_transform(x_train), columns=x_train.columns, index = x_train.index)
x_test = pd.DataFrame(scalar.fit_transform(x_test), columns=x_test.columns, index = x_test.index)
```

```python
x_train
```

```python
x_test.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x_train, y_train)
```

```python
print(model.feature_importances_)
```

```python
feat_importances = pd.Series(model.feature_importances_,index=x_train.columns)
feat_importances.nlargest(8).plot(kind='barh')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
print("Coefficient of determination R^2 <-- on train set {}".format(regressor.score(x_train,y_train)))
print("Coefficient of determination R^2 <-- on test set {}".format(regressor.score(x_test,y_test)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
print("Coefficient of determination R^2 <-- on train set {}".format(rf.score(x_train,y_train)))
print("Coefficient of determination R^2 <-- on test set {}".format(rf.score(x_test,y_test)))
```

```python
y_pred = regressor.predict(x_test)
```

```python
sns.distplot(y_test-y_pred)
```

```python
sns.scatterplot(y_test,y_pred)
```

```python
y_predict = rf.predict(x_test)
```

```python
sns.distplot(y_test-y_predict)
```

```python
sns.scatterplot(y_test, y_predict)
```

```python
from sklearn import metrics
print('MAE :', metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
from sklearn import metrics
print('MAE :', metrics.mean_absolute_error(y_test,y_predict))
print('MSE:', metrics.mean_squared_error(y_test,y_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
